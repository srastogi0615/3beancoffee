```python
import pandas as pd
url = '..\SeasonStats.csv'
seasons = pd.read_csv(url, index_col='Rk')
seasons.head()
```

```python
seasons = seasons.drop('MOV', 1)
seasons = seasons.drop('SOS', 1)
seasons.head()
```

```python
seasons['Tm'] = seasons['Tm'].map(lambda x: x.rstrip('*'))
seasons['Tm']
```

```python
url2 = '..\Champions.csv'
champions = pd.read_csv(url2)
champions.head()
```

```python
seasons['team_year'] = list(zip(seasons.Tm, seasons.Champ_Year))
seasons.head()
```

```python
champions['winner_year'] = list(zip(champions.Champ_Ack, champions.Year))
champions['runner_year'] = list(zip(champions.Runner_Ack, champions.Year))
champions.head()
```

```python
seasons['iswinner'] = seasons['team_year'].map(champions.set_index('winner_year')['Year'])
seasons['isrunner'] = seasons['team_year'].map(champions.set_index('runner_year')['Year'])
```

```python
seasons.head()
```

```python
seasons.loc[seasons['iswinner'].notnull(), 'iswinner'] = 1
```

```python
seasons.head()
```

```python
seasons.loc[seasons['isrunner'].notnull(), 'isrunner'] = 1
```

```python
seasons.head()
```

```python
seasons['iswinner'].fillna(0, inplace=True)
```

```python
seasons.head()
```

```python
seasons['isrunner'].fillna(0, inplace=True)
seasons.head()
```
