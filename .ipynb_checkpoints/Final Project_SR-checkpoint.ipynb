# Project Overview:
The ago old question when it comes to success in almost all team sports boils down to the same central dichotomy; is offense or defense they key to championship contention. The common phrase heard in football and basketball is that 'defense wins championships' and well respected coaches around the league value players and styles that optimize team defensive potential. In this project, I am attempting to use data science techniques to explore this central dichotomy for prediciting if NBA teams will win the championship. The goal of the project is to use historical analysis to see if team compositions and play styles are better set to be offensive/defensive-minded with the end goal being winning the NBA championship.

## Data Description
I chose to explore NBA Advanced Statistics from basketball-reference.com.  These Advanced Statistics control for differences in individual game flow (see defintions below).  I scraped the last 30 regular seasons for all NBA teams from basketball-reference. Some of the key statistics I knew I wanted to explore are defined below.

- SRS (Simple Rating System): A team rating that takes into account average point differential and stregth of schedule
- Pace: An estimate of possessions per 48 minutes
- ORtg (Offensive Rating): An estimate of points produced per 100 possessions
- DRtg (Defensive Rating): An estimate of points allowed per 100 possessions
- eFG% (Effective Field Goal Percentage): Field goal percentage adjusting for the fact that 3pt field goal is worth more than a 2-pt field goal
- Opponent eFG%: Team's opponent's EFG

The way I classified these based on definition are that ORtg and eFG% are offensive indicators and DRtg and Opponent eFG% are defensive indicators, while SRS and Pace are neutral or independent of offensive and defensive indicators.

## Model 1: KNN

The first model I used to predict the NBA champion based off of Advanced Statistics was KNN classification. The model with the ranked stats as features ended up having the same accuarcy as the null case (0.964).  As such, I chose to tune the n_neighbors for models using the absolute stats.  The optimal value of k was found to be 30 however all the feature combinations I created using this classifier resulted in the same accuarcy score (0.965) 50 bps better than the null model. Though performance for this model was solid, it did not really tell me which features are more important towards predicting a championship contender.

```python
# Creating Feature Maxtrix and Response Vector
feature_cols_rnk = ['SRS_Rank', 'Pace_Rank', 'ORtg_Rank','DRtg_Rank', 'eFG%_Rank', 'OeFG%_Rank']
feature_cols = ['SRS', 'Pace', 'ORtg', 'DRtg', 'eFG%-T', 'eFG%-O']
y = seasons.iswinner
```

```python
#KNN model evaluated via traintestsplit and rank columns as features
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
X = seasons[feature_cols_rnk]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=99)
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train, y_train)
y_pred_class = knn.predict(X_test)
print metrics.accuracy_score(y_test, y_pred_class)
```

```python
#compare with null accuracy
y_test.value_counts()
```

```python
y_test.value_counts().head(1) / len(y_test)
```

```python
#KNN model evaluated via traintestsplit and absolute stats as features
X = seasons[feature_cols]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=99)
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train, y_train)
y_pred_class = knn.predict(X_test)
print metrics.accuracy_score(y_test, y_pred_class)
```

```python
#KNN with cross validation absolute stats
from sklearn.model_selection import cross_val_score
k_range = range(1, 51)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
    k_scores.append(scores.mean())
print k_scores
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
```

```python
#KNN with cross validation ranked stats
X = seasons[feature_cols_rnk]
knn = KNeighborsClassifier(n_neighbors=30)
scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
print scores.mean()
```

## Model 3: KNN Clustering

In order to see if there were groups or clusters of championship winning teams with similar advanced stastics, I created a new dataframe 'winseasons' that filtered seasons when 'iswinner' equaled 1. From this dataframe I started by testing to see whether the absolute statistics (feature_cols) or the ranked statistics (feature_cols_rnk) produced better (tighter) clusters. The absolute statistics did show to have a better fitted scatterplot than the ranked stats and as such was used for the tuning portion of the clustering.

I tested different combinations of feature sets to maximize the silhouette coefficient and determine the optimal # of clusters. The top two feature sets were feature_cols2 (ORtg, DRtg, eFG%-T, eFG%-O) with 0.5 and feature_cols4 with 0.7 (DRtg, eFG%-T).  However, the optimal number of clusters for feature_cols4 set was 8 versus 2 for feature_cols2. To reduce the complexity, I decided to move forward with feature_cols2 clusters for further exploration.

From analyzing the clusters mean values for the advanced stats they applied for, I found that the first cluster (cluster 0) stressed more importancee towards offense with ORtg and eFG%-T with means of 3.9 and 4.3 vs. 9.5 and 9.1 for the second cluster. The second cluster (cluster 1) was skewed more heavily towards defense with DRtg and eFG%-O with means of 2.0 and 2.6 vs. 6.5 and 7.0 for the first cluster. With this clear delineation between the clusters, I analyzed the counts of the clusters and found that 21, or 68%, of the last 31 championship winning teams were part of cluster 0 and that 8 out of the last 10 champions were in cluster 0 with the exception being the 2008 Boston Celtics and the 2012 Miami Heat. 4 out of the 10 teams in cluster were San Antonio Spurs teams and the other notable team was the 2004 Detroit Pistons, a team heralded for their defenstive prowess.

## Model 5: Random Forests
To control for the variance of the single decision tree classifer, I tested a Random Forests model. Tuning the model for the offensive and defensive statistics resulted in optimal n_estimators=10 and max_features=2.  Though the accuracy matched closely to the null, the feature importance rankings revealed once again that ORtg was most important followed by DRtg, eFG%-O, and least important eFG%-T.

```python
#Random Forest model
from sklearn.ensemble import RandomForestClassifier
cfreg = RandomForestClassifier()
cfreg
```

```python
cfreg.fit(X, y)
```

```python
score = cross_val_score(cfreg, X, y, cv=10, scoring='accuracy')
score.mean()
```

```python
#Tuning n-estimators
estimator_range = range(10, 310, 10)
accuracy_scores = []
for estimator in estimator_range:
    cfreg = RandomForestClassifier(n_estimators=estimator, random_state=1)
    scores = cross_val_score(cfreg, X, y, cv=10, scoring='accuracy')
    accuracy_scores.append(np.mean(scores))
```

```python
plt.plot(estimator_range, accuracy_scores)
plt.xlabel('n_estimators')
plt.ylabel('accuracy (higher is better)')
```

```python
sorted(zip(accuracy_scores, estimator_range), reverse=True)[0]
```

```python
#Tuning max features
feature_range = range(1, len(feature_cols2)+1)
accuracy_scores = []
for feature in feature_range:
    cfreg = RandomForestClassifier(n_estimators=10, max_features=feature, random_state=1)
    scores = cross_val_score(cfreg, X, y, cv=10, scoring='accuracy')
    accuracy_scores.append(np.mean(scores))
```

```python
plt.plot(feature_range, accuracy_scores)
plt.xlabel('max_features')
plt.ylabel('accuracy (higher is better)')
```

```python
sorted(zip(accuracy_scores, feature_range), reverse=True)[0]
```

## Conclusion
Though the Advanced Statistics did not prove to have much value in predicting the NBA champsionship over the null (random) use case, the importance and clusters of features did lead to some interesting insights on the central question: Is offense or defense more important to winning the championship. It appears that there is a higher probability of success if the team is more offensively focused. This is supported by the fact that KNN clustering led to a two-cluster model which had 21 out of the last 31 winning championship teams grouped in the offensive stat skewed cluster and Random Forest feature ranking had ORtg as the most important feature. This does hold true to the pundits discussion around how the modern league with its rule changes and increased reliance on the 3pt shot is better suited for offensive minded teams (10 out of the last 12 champions were in the offesnive cluster). 

In general, however both offensive and defensive prowess will likely lead to a successful scenario. It is interesting that the more defensive minded teams had a lower mean rank for the defensive stats than the offensive minded championship teams did for the offesnive stats. Perhaps this only further reduces the probability that a defensive first approach can lead to a championship success as a defensive approach could require more extreme prowess on that end. The best example that illustrates the use case of 'Defense is the key to Championships' is th San Antonio Spurs. In, 4 out of their 5 championship seasons they had better defensive rankings than offensive rankings. The Spurs are renowned for their team culture, personnel continuity (in coaching and players), and commitment to team philosophy. For teams looking to reach that top status, it could be hard to develop such a culture as the San Antonio Spurs have overnight. However, if it is possible, perhaps they too can be a multi-winning championship team that puts their franchise in the NBA recordbooks.
